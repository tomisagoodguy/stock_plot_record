```python
def check_signals(stock_data):
    # 計算波動率（使用日收益率的標準差）
    volatility = stock_data['returns'].rolling(window=20).std()

    # 計算成交量變動率
    volume_change = stock_data['volume'].pct_change()

    # 設定警戒標準
    vol_threshold = volatility.mean() + volatility.std()
    volume_drop_threshold = -0.2  # 成交量下降20%

    # 檢查訊號
    high_vol = volatility > vol_threshold
    low_volume = volume_change < volume_drop_threshold

    return high_vol, low_volume
```
